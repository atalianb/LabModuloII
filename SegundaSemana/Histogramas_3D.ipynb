```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
N = 1000#Numero de puntos
L = 100#Longutud de la caja
data = L*np.random.rand(N,3)#Genera un arreglo de  N puntos aleatorios con 2 dim
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data.T[0,:],data.T[1,:],data.T[2,:])
```

```python
def distancia(x1,x2,y1,y2,z1,z2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2 + (z2-z1)**2)
    return d
```

```python
ran = L*np.random.rand(N,3)
```

```python
dis_data = []
dis_ran = []
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(data[i][0],data[j][0],data[i][1],data[j][1],data[i][2],data[j][2])
        d_r = distancia(ran[i][0],ran[j][0],ran[i][1],ran[j][1],ran[i][2],ran[j][2])
        dis_data.append(d_d)
        dis_ran.append(d_r)
```

```python
data_n, data_bin,m = plt.hist(dis_data,bins=20)
```

```python
ran_n, ran_bin, p = plt.hist(dis_ran,bins=20)
```
