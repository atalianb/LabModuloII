```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import scipy.fftpack
%matplotlib inline
```

Usando $X_{k} = \sum^{N-1}_{n = 0} = x_{n} e^{-2\pi ikn/N}$
donde
N = número de muestras
n = muestra actual
$x_{n}$ = valor de la señal en el tiempo n
k = frecuencia
$X_{k}$ = Discret Fourier Transform

```python
def DFT(x):#x es un arreglo de dimensión 1
    N = x.size
    n = np.arange(N)
    k = n.reshape((N,1))
    expo = np.exp(-2j*np.pi*k*n/N)
    result = np.dot(expo,x)
    return result
```

```python
t = np.linspace(0, 2*np.pi, 1000, endpoint=True)
f = 3.0 # Frecuencia en Hertz
A = 100.0 # Amplitud
s = A * np.sin(2*np.pi*f*t) #señal
```

```python
plt.plot(t,s)
plt.xlabel('tiempo')
plt.ylabel('Aamplitud')
plt.show()
```

```python
F = DFT(s)
plt.plot(F)
plt.show()
```

Para comprobar que estamos haciendo bien la transformada de Fourier usamos la función de numpy

```python
Y = np.fft.fft(s)
plt.plot(Y)
plt.show()
```
