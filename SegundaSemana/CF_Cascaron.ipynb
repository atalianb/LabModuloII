```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ran_x = np.random.uniform(0.,1.4,1000)
ran_y = np.random.uniform(0.,1.4,1000)
```

```python
plt.plot(ran_x,ran_y,'.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
theta = np.random.random(1000)*2.*np.pi
r1 = np.random.uniform(0.4,0.5,1000)
r = np.sqrt(r1)#**(0.5)
x = r*np.cos(theta) +0.7
y = r*np.sin(theta) + 0.7
plt.plot(x,y,'.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(ran_x,ran_y,'.')
plt.plot(x,y,'.')
plt.show()
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dis_data = []
dis_ran = []
```

```python
for i in range(len(x)):
    for j in range(i+1,len(x)):
        d_d = distancia(x[i],x[j],y[i],y[j])
        d_r = distancia(ran_x[i],ran_x[j],ran_y[i],ran_y[j])
        dis_data.append(d_d)
        dis_ran.append(d_r)
```

```python
data_n, data_bin,m = plt.hist(dis_data,bins=20)
```

```python
ran_n, ran_bin, p = plt.hist(dis_ran,bins=20)
```

Funcin de Correlacin de Peebles $\frac{DD(r)}{RR(r)} - 1$ 

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1)
plt.ylim(-1.,10.)
plt.show()
```

Para la funcin de correlacin de Hamilton $\psi_{H}^{2} = \frac{DD(r)RR(r)}{(DR(r))^{2}}$

```python
dis_dr = []
```

```python
for i in range(len(x)):
    for j in range(len(x)):
        d_dr = distancia(x[i],ran_x[j],y[i],ran_y[j])
        dis_dr.append(d_dr)
```

```python
dr_n, dr_bin, q = plt.hist(dis_dr,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(dr_n)**2)
plt.show()
```

Para la funcin de correlacin de Landy-Szalay $1 + \frac{DD(r)}{RR(r)} -2\frac{DR(r)}{RR(r)} $

```python
plt.plot(data_bin[1:],1.+ data_n/ran_n -2*dr_n/ran_n)
plt.ylim(-15.,10.)
plt.show()
```
