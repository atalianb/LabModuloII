```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 1000#Numero de puntos
L = 100#Longutud de la caja
data = L*np.random.rand(N,2)#Genera un arreglo de  N puntos aleatorios con 2 dim
```

```python
plt.plot(data,'.')
```

```python
ran = L*np.random.rand(N,2)
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dtot = N*(N-1)/2
j = 0
dold = 0
dmax = 0
dis_data = []
dis_ran = []
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(data[i][0],data[j][0],data[i][1],data[j][1])
        d_r = distancia(ran[i][0],ran[j][0],ran[i][1],ran[j][1])
        dis_data.append(d_d)
        dis_ran.append(d_r)
```

```python
data_n, data_bin,m = plt.hist(dis_data,bins=20)
```

```python
ran_n, ran_bin, p = plt.hist(dis_ran,bins=20)
```

Funcin de Correlacin de Peebles $\frac{DD(r)}{RR(r)} - 1$ donde en este caso estamos tomando a los datos como puntos aleatorios tambin

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1)
```

Para la funcin de correlacin de Hamilton $\psi_{H}^{2} = \frac{DD(r)RR(r)}{(DR(r))^{2}}$

```python
dis_dr = []
```

```python
for i in range(N):
    for j in range(N):
        d_dr = distancia(data[i][0],ran[j][0],data[i][1],ran[j][1])
        dis_dr.append(d_dr)
```

```python
dr_n, dr_bin, q = plt.hist(dis_dr,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(dr_n)**2)
```

Para la funcin de correlacin de Landy-Szalay $1 + \frac{DD(r)}{RR(r)} -\frac{DR(r)}{RR(r)} $

```python
plt.plot(data_bin[1:],1.+ data_n/ran_n -dr_n/ran_n)
```
