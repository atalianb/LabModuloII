```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from scipy.special import erf
%matplotlib inline
```

```python
rand_arr = np.random.randn(1000)
```

```python
n = 1000#número de pasos
x = np.zeros(n)#Arreglo de ceros con n elementos
cdf = np.zeros(n)
```

```python
mu = 70.
sigma = 2.
```

```python
x = np.random.normal(mu, sigma, n)
#print x
xprom = np.mean(x)
```

```python
#Función de distribución acumutariva (CDF) para una gaussiana
cdf = 0.5*(1.0 + erf((x-xprom) / (sigma*sqrt(2.0))))
```

```python
plt.scatter(x,cdf,marker='_')
plt.show()
```

```python
plt.hist(x, bins=50, normed=True) 
plt.show()
```
